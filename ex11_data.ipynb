```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 200)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 200)
y = np.cos(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 200)
y = np.tan(x)
plt.plot(x,y)
plt.show()
```
